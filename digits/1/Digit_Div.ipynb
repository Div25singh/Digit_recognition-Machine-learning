```python
import numpy as py
import pandas as pd
from sklearn.linear_model import LinearRegression
clf1=LinearRegression()
```

```python
x=pd.read_csv("C:/Users/500062812/Desktop/digits/digits/train.csv")
y=pd.read_csv("C:/Users/500062812/Desktop/digits/digits/test.csv")
```

```python
print(x.tail())
```

```python
print(y.tail())
```

```python
train_data = pd.DataFrame(x)
test_data = pd.DataFrame(y)
```

```python
dtrainx=train_data.iloc[:,1:]
dtrainy=train_data.iloc[:,0:1]
```

```python
print(dtrainx.tail())
print(dtrainy.tail())
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dtrainx, dtrainy, test_size = 0.4, random_state = 1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
clf1.fit(x_train,y_train)
```

```python
print(clf1.intercept_)
print(clf1.coef_)
```

```python
#y_test = pd.DataFrame(y_test)
#_test=py.around(y_test,decimals=1)
#_test=pd.DataFrame(y_test)
#y_test
```

```python
pred=clf1.predict(x_test)
pred=(py.around(pred,decimals=0))
print(pd.DataFrame(pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(pred, y_test)
```

```python
import matplotlib.pyplot as plt
c = [i for i in range(0,16800,1)]         # generating index 
fig = plt.figure()
plt.plot(c,y_test, color="blue", linewidth=2.5, linestyle="-")
plt.plot(c,pred, color="red",  linewidth=2.5, linestyle="-")
fig.suptitle('Actual and Predicted', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                               # X-label
plt.ylabel('Height', fontsize=16) 
```

```python
c = [i for i in range(0,16800,1)]
fig = plt.figure()
plt.plot(c,y_test-pred, color="blue", linewidth=2.5, linestyle="-")
fig.suptitle('Error Terms', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                      # X-label
plt.ylabel('ytest-ypred', fontsize=16)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, pred)
r_squared = r2_score(y_test, pred)
```

```python
print('Mean Squared Error :' ,mse)
print('r_square_value :',r_squared)
print('RMSE :',py.sqrt(mean_squared_error(y_test,pred)))
```
